```python
#OUR CONTRIBUTION
def pos_str_retrieve(sent):
  wordlist = nltk.word_tokenize(sent)
  tagged = nltk.pos_tag(wordlist)
  pos_str = ""
  for tag in tagged:
    pos_str = pos_str + tag[1] + " "
  return pos_str.rstrip(' ')
```

```python
#OUR CONTRIBUTION
def pos_str_retrieve2(sent):
  wordlist = nltk.word_tokenize(sent)
  tagged = nltk.pos_tag(wordlist)
  pos_str = ""
  for tag in tagged:
    pos_str = pos_str + tag[1] + " "
  return pos_str.rstrip(' '),tagged
```

```python
#OUR CONTRIBUTION
def retag(stringtree):
  #this function helps in removing tenses related info and streamline parsetrees for comparison
  #types of verbs
  stringtree = stringtree.replace("VBG","VB")
  stringtree = stringtree.replace("VBD","VB")
  stringtree = stringtree.replace("VBN","VB")
  stringtree = stringtree.replace("VBP","VB")
  stringtree = stringtree.replace("VBZ","VB")
  #types of nouns
  stringtree = stringtree.replace("NNS","NN")
  stringtree = stringtree.replace("NNP","NN")
  stringtree = stringtree.replace("NNPS","NN")
  #types of adverbs
  stringtree = stringtree.replace("RBR","RB")
  stringtree = stringtree.replace("RBS","RB")
  #types of adjectives
  stringtree = stringtree.replace("JJR","JJ")
  stringtree = stringtree.replace("JJS","JJ")

  return stringtree
```
